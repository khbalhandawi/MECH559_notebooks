# Optimization of a cylindrical surface area

Consider the following cylinder with radius $r$ and height $h$

<p style="text-align:center;">
    <img src="./cylinder.jpg" alt="beam 1" title="A cantilever beam" width="350px" align="center"/>
</p>

We wish to maximize the surface area subject to a volume constraint

\begin{equation*}
	\begin{aligned}
		& \underset{r,h}{\text{minimize}}
		& & f(r,h) = -2\pi r^2 - 2\pi r h \\
		& \text{subject to}
		& & g(r,h) = \pi r^2 h -1 \leq 0\\
	\end{aligned}
\end{equation*}

From monotonicity we know that g(r,h) is active: $$ f(r^-,h^-) = -2\pi r^2 - 2\pi r h$$ and $$ g(r^+,h^+) = \pi r^2 h -1$$

We can transform the constrained optimization problem to an unconstrained problem through substitution of the active constraint 
$$g(r,h) = \pi r^2 h -1 = 0 \rightarrow h = \dfrac{1}{\pi r^2}$$
the optimization problem becomes

\begin{equation*}
	\begin{aligned}
		& \underset{r}{\text{minimize}}
		& & f(r) = -2 \pi r^2 - \dfrac{2}{r} \\
	\end{aligned}
\end{equation*}

However, this is not a well-posed problem. Why?

```julia
using Plots

# independant parameter
r = LinRange(-1,1,100)

# dependant paramter
f(r) = -2**(r^2) - (2/r)

plot(r,f,label=nothing)
xlabel!("r")
ylabel!("f(r)")
```
