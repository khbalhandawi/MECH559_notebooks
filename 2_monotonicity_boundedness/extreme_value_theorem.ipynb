# Mathematical preliminaries

## The norm of a vector

let $\mathbf{x}\in\mathbb{R}^n$, then $\left|\left|\mathbf{x}\right|\right|$ (norm of $\mathbf{x}$) is any function which statisfies

* $\left|\left|\mathbf{x}\right|\right|=0$ iff $\mathbf{x} = \mathbf{0}$
* $\left|\left|\alpha\mathbf{x}\right|\right|~=~\left|\alpha\right|\left|\left|\mathbf{x}\right|\right|$ $\forall \alpha \in \mathbb{R}$
* $\left|\left|\mathbf{x}+\mathbf{y}\right|\right| \leq \left|\left|\mathbf{x}\right|\right| + \left|\left|\mathbf{y}\right|\right|$ (triangle inequality)

The norm is a measure of a vector's length. There are different types of norms

* 2-norm (euclidean distance or $\ell 2$ norm):
$\left|\left|\mathbf{x}\right|\right|_2=\sqrt{x_1^2 + \cdots + x_n^2}$

* 1-norm:
$\left|\left|\mathbf{x}\right|\right|_1=\sum_{i=1}^{n} \left|x_i\right|$

* $\infty$-norm:
$\left|\left|\mathbf{x}\right|\right|_\infty= \underset{i}{\mathrm{max}} \left|x_i\right|$

* $p$-norm:
$\left|\left|\mathbf{x}\right|\right|_p= \left(\left|x_1\right|^p + \cdots + \left|x_n\right|^p\right)^{\frac{1}{p}}$

As $p \rightarrow \infty$ weget the $\infty$-norm from the p-norm. 

Unless otherwise it is explicilty stated, we will use $\left|\left|\mathbf{x}\right|\right|$ to denote $\left|\left|\mathbf{x}\right|\right|_2$

## Continuity of a function

A function $f$ is said to be continuous at a point $\mathbf{x}$ if

* given $\varepsilon>0$ $\exists~\delta > 0$ s.t. $$\left|f(\mathbf{x}) - f(\mathbf{y})\right| \leq \varepsilon~\forall~\left|\left|\mathbf{y} - \mathbf{x} \right|\right| \leq \delta$$, or equivalently

* $\underset{i\rightarrow \infty}{\text{lim}} f(\mathbf{x}_i) = f(\mathbf{x})$ if $\mathbf{x}_i \rightarrow \mathbf{x}$

if $f$ is continuous $\forall \mathbf{x} \in \mathbb{R}^n$ (at all points $\mathbf{x}$) then $f$ is said to be continuous.

For example:

the functions
$$
    f_1(x)={\begin{cases}
        x   & {x \leq 0}\\
        x+1 & {x > 0}
    \end{cases}}
$$ and
$$
    f_2(x)=\dfrac{1}{x - 2}
$$
are non-continuous on $x\in\mathbb{R}$

```julia
using Plots,LaTeXStrings
# independent parameters
x = LinRange(-4,4,1000)
x₁ = LinRange(-4,2,100)
x₂ = LinRange(2,4,100)

# dependant variables
f₁(x) = if x <= 0 x else x+1 end
f₂(x) = 1 / (x-2)

# plot
plot(x, f₁, lw = 1, label=L"f_1(x)")
plot!(x₁, f₂, lw = 1, color = "orange", label=L"f_2(x)")
plot!(x₂, f₂, lw = 1, color = "orange", label=nothing)
xlabel!("x")
ylabel!("y")
ylims!((-2,3))
# savefig(p1,"active_constraint.pdf")
```

## closed sets and open sets

* A set $\mathcal{S} \in \mathbb{R}$ is said to be open if we can draw a small ball around each $x\in \mathcal{S}$ such that the ball $\subseteq \mathcal{S}$.\
i.e., $\exists~\varepsilon>0$ s.t.
$$ \left\{y:\left|\left|\mathbf{x} - \mathbf{y}\right|\right| \leq \varepsilon\right\} \subseteq \mathcal{S} $$

* $\mathcal{S}$ is closed if its compliment $\mathcal{S}'$ is open

examples:

* $x>0$ is an open set
* $x\geq 0$ is a closed set
* $\dfrac{1}{2} < x\leq 1$ is neither closed or open

Compact sets: A set $\mathcal{S} \subseteq \mathbb{R}^n$ is said to be compact if it is closed and *bounded*, i.e., it is closed and $\exists~M < \infty$ s.t
$$ \left|\left|\mathbf{x}\right|\right| \leq~M~\forall~\mathbf{x} \in \mathcal{S}$$

# Extreme value theorem

The extreme value theorem (also known as Weierstrass' extreme value theorem) states that:

Let $f$ be a **continuous function** over a **compact** set $\mathcal{S}$. Then $f$ achieves both a minimum and a maximum over $\mathcal{S}$. i.e., $\exists~x_1$ s.t.

\begin{align*}
	& f(x_1) = \underset{x \in \mathcal{S}}{\text{infimum}}
	& & f(x)
\end{align*}

and $\exists~x_2$ s.t.

\begin{align*}
	& f(x_2) = \underset{x \in \mathcal{S}}{\text{supremum}}
	& & f(x)
\end{align*}

Corollary:

* if $f$ is continuous and $f(x) \rightarrow \infty$ as $\left|\left|x\right|\right| \rightarrow \infty$ then a global minimum exists.
* if $f$ is continuous and $f(x) \rightarrow -\infty$ as $\left|\left|x\right|\right| \rightarrow \infty$ then a global maximum exists

This theorem is an indicator of a well-posed optimization problem.

## example 2: infimum, supremum, max, and min of a 2D function

(a) Find the inf and sup of the function

$$
f(x_1,x_2) = x_1 + 2x_2
$$

over the set 

$$
\mathcal{D} = \left\{x:x_1^2 + x_2^2 \le 1, x_1>0, x_2>0\right\}
$$

---

We examine the 2D function by plotting its isocontours and the set $\mathcal{D}$. Note that dotted lines $\not\in \mathcal{D}$

```julia
# independent parameters
x1 = LinRange(-0.1,1.2,100)
x2 = LinRange(-0.1,1.2,100)

# dependant variables
f(x1,x2) = (x1) + 2*(x2)

# constraint function
y_cstr(x1) = sqrt(1 - x1^2)

# optimum isocontours
y_iso(x1) = (sqrt(5) - x1) / 2
y_iso_perpindicular(x1) = 2*x1

# maximum
x_sup = [1/sqrt(5),2/sqrt(5)]
x_inf = [0,0]

# active constraint
p1 = contour(x1, x2, f, fill = false, levels = 5, color = "black", colorbar = false, contour_labels=true, size=(600,600))
plot!(LinRange(0,1,100), y_cstr, lw = 2, label=L"{x_1}^2 + {x_2}^2 \leq 1")
plot!([0,1], [0,0], lw = 2, linestyle=:dash, label=L"x_2 > 0")
plot!([0,0], [0,1], lw = 2, linestyle=:dash, label=L"x_1 > 0")
plot!(LinRange(0,1,100), y_iso, lw = 3, label=L"\mathrm{least~upper~bound:}~x_1 + 2x_2 = \sqrt{5}")
plot!(LinRange(0,x_sup[1],100), y_iso_perpindicular, lw = 3, label=L"\mathrm{perpindicular:}~2x_1 - x_2 = 0")
scatter!([x_sup[1]], [x_sup[2]], markersize = 7, label=L"\mathbf{x}_\mathrm{sup}")
scatter!([x_inf[1]], [x_inf[2]], markersize = 7, label=L"\mathbf{x}_\mathrm{inf}")
xlabel!(L"x_1")
ylabel!(L"x_2")
ylims!((-0.1,1.2))
xlims!((-0.1,1.2))
# savefig(p1,"active_constraint.pdf")
```

Consider the isocontours $x_1+2x_2 = c$. The line with the biggest value $c$ (least upper bound) is tangent to the circle 
\begin{equation}
\tag{1}
x_1^2 + x_2^2 = 1
\end{equation}
The radial vector starting the center of the circle $(0,0)$ is perpendicular to this tangent. 

The slope of the line for this vector is $m = 2$, since the slope of $x_1+2x_2 = c \rightarrow x_2 = \dfrac{c}{2} - \dfrac{x_1}{2}$ is $m = -\dfrac{1}{2}$. 

The y-intercept is $b=0$ and the equation of the perpendicular line becomes:

\begin{equation}
\tag{2}
x_2 = 2x_1
\end{equation}

Solving Equations (1) and (2) yields

$$\mathbf{x}_\text{sup} = \left(\dfrac{1}{\sqrt{5}},\dfrac{2}{\sqrt{5}}\right)$$

This is shown on the figure above.

The infimum occurs at 

$$\mathbf{x}_\text{inf} = \left(0,0\right)$$

---

(b) Does a (i) minimum and a (ii) maximum exist for this problem?

---

(i) No minimum exists for this problem since $\left(0,0\right) \not\in \mathcal{D}$

(ii) The maximum is equal to the supremum since the points $\left\{x_1,x_2:x_1^2 + x_2^2 = 1\right\} \in \mathcal{D}$

## example 3: a 3D function

Does the following function 

$$
f(\mathbf{x}) = x_1^2 + \exp{\left(x_2\right)} + \exp\left(-x_2\right) + 3x_3^4 
$$

over the set

$$
\mathcal{D} = \left\{\mathbf{x}:x_1^2 + 2x_2^2 + 3x_3^2 \leq 1\right\}
$$

have

(i) a minimum?

(ii) a maximum? 

---
Both a minimum and maximum exist for this function by the extreme value theorem, since $f$ is continuous and $\mathcal{D}$ is a compact set. We do do not need to solve the problem to show that a solution exists

We visualize the ellipsoids that describes the boundary of the set $\mathcal{D}$ below

```julia
pyplot()
using LinearAlgebra

# https://discourse.julialang.org/t/drawing-an-ellipsoid-to-visualize-a-tensor/31286/2?u=khbalhandawi
function ellipsoid(center, coefs, ngrid=25)
    # Radii corresponding to the coefficients:
    rx, ry, rz = 1 ./ sqrt.(coefs)

    # Set of all spherical angles:
    u = range(0, 2pi, length=ngrid)
    v = range(0, pi, length=ngrid)

    # Cartesian coordinates that correspond to the spherical angles:
    # (this is the equation of an ellipsoid):
    x = [rx * x * y for (x, y) in  Iterators.product(cos.(u), sin.(v))]
    y = [ry * x * y for (x, y) in Iterators.product(sin.(u), sin.(v))]
    z = [rz * x * y for (x, y) in Iterators.product(ones(length(u)), cos.(v))]
    return x .+ center[1], y .+ center[2], z .+ center[3]
end

# Plot:
A = [1 2 3] # ellipse coeffs
C = [0 0 0] # ellipse center

x, y, z = ellipsoid(C, A)
# surface(x, y, z, alpha=0.5)
p1 = plot(x,y,z,st=:surface,camera=(-30,30), alpha=0.5, colorbar=false, 
    xlabel = L"x_1", ylabel = L"x_2", zlabel = L"x_3")
ylims!((-1.2,1.2))
xlims!((-1.2,1.2))
zlims!((-1.2,1.2))
```
